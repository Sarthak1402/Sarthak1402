# Required Libraries and Dataset

```python
# Libraries required

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data

Ice_cream_review = {
    'Day' : ['1', '2', '3', '4', '5', '6', '7'],
    'Flavor Rating' : [0.22, 0.63, 0.44, 0.39, 0.37, 0.88, 0.14],
    'Texture Rating' : [0.40, 0.94, 0.04, 0.54, 0.52, 0.19, 0.35],
    'Overall Rating' : [0.60, 0.11, 0.60, 0.49, 0.99, 0.83, 0.10]
}
```

```python
# Converting into pandas dataframe

df = pd.DataFrame(Ice_cream_review)
df
```

```python
# Setting the index

df.set_index('Day', inplace = True)
```

# Line Plot

***Line Plot Syntax: df.plot(dataframe, kind, x_label, y_label, subplots)***

```python
# Simple line Plot

df.plot(kind = 'line')
```

```python
# Line plot but seperated by columns

df.plot(kind = 'line', subplots = True)
```

```python
# Adding title & labels

df.plot(kind = 'line', title = 'Ice Cream', xlabel = 'Review Day', ylabel = 'Rating')
```

# Bar Plot

```python
# Simple Bar Plot

df.plot(kind = 'bar')
```

```python
# Stacked Bar plot

df.plot(kind = 'bar', stacked = True)
```

```python
# Bar plot on specific columns

df['Flavor Rating'].plot(kind = 'bar')
```

```python
# Horizontal bar plot

df['Flavor Rating'].plot(kind = 'barh')
```

```python
# Sorted Horizontal Bar Plot

df2 = df.sort_values(by = 'Flavor Rating')

df2['Flavor Rating'].plot(kind = 'barh')
```

# Scatter Plot

***Syntax: df.plot.scatter(x = 'column_1', y = 'column_2')***

```python
# Simple Scatter Plot

df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating')
```

```python
# Changing size and colout of data points

df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 300, c = 'Red')
```

# Histogram

***Syntax: df.plot.hist(bins)***

```python
# Simple Histogram

df['Flavor Rating'].plot.hist(bins = 6)
```

# Boxplots

***Syntax: df.boxplot()***

```python
# Simple Box plots

df.boxplot()
```

```python
# Adjusting figure size

df.boxplot(figsize = (6,4))
```

# Area Plot

***Syntax: df.plot.area()***

```python
df.plot.area(figsize = (10,5))
```

# Pie Chart

***Syntax: df.plot.pie(y = 'column_name')***

```python
df.plot.pie(y = 'Flavor Rating')
```
